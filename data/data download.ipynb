```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 25000   # 20k–30k rows

df = pd.DataFrame({

    "LotArea": np.random.randint(2000, 15000, n),
    "GrLivArea": np.random.randint(500, 4000, n),
    "GarageArea": np.random.randint(0, 1000, n),
    "OverallQual": np.random.randint(1, 10, n),
    "OverallCond": np.random.randint(1, 9, n),
    "YearBuilt": np.random.randint(1950, 2022, n),
    "YearRemodAdd": np.random.randint(1950, 2022, n),
    "TotalBsmtSF": np.random.randint(0, 2000, n),
    "FullBath": np.random.randint(1, 4, n),
    "HalfBath": np.random.randint(0, 2, n),
    "TotRmsAbvGrd": np.random.randint(3, 12, n),
    "Fireplaces": np.random.randint(0, 3, n),
    "GarageCars": np.random.randint(0, 4, n),

    # categorical features
    "Neighborhood": np.random.choice(["NAmes","CollgCr","OldTown","Edwards","Somerst"], n),
    "MSZoning": np.random.choice(["RL","RM","FV","RH"], n),
    "BldgType": np.random.choice(["1Fam","2fmCon","Duplex"], n),
    "HouseStyle": np.random.choice(["1Story","2Story","1.5Fin"], n),
    "KitchenQual": np.random.choice(["TA","Gd","Ex"], n),
    "ExterQual": np.random.choice(["TA","Gd","Ex"], n),
    "Foundation": np.random.choice(["PConc","CBlock","BrkTil"], n),
    "CentralAir": np.random.choice(["Y","N"], n),
    "PavedDrive": np.random.choice(["Y","N"], n),
    "HeatingQC": np.random.choice(["Ex","Gd","TA"], n),
})

# 🎯 realistic price formula (IMPORTANT)
df["SalePrice"] = (
    df["GrLivArea"] * 120 +
    df["OverallQual"] * 15000 +
    df["GarageCars"] * 8000 +
    df["TotalBsmtSF"] * 50 +
    np.random.randint(-20000, 20000, n)
)

# save
df.to_csv("houses.csv", index=False)

print("✅ 25,000 rows realistic dataset generated")
print(df.head())
```
